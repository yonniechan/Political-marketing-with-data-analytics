```python
import pandas as pd
# import spacy
import re
# import string
from collections import Counter
from spacy.lang.en.stop_words import STOP_WORDS
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
tweets = pd.read_csv('tweets.csv')
```

```python
trump = tweets[tweets['handle'] == 'realDonaldTrump']
clinton = tweets[tweets['handle'] == 'HillaryClinton']
```

```python
trump_text = ''
trump_text = trump['text'].str.cat(sep=' ')
trump_text = re.sub(r'[^\w\s]', '', trump_text).lower()
```

```python
trump_words = trump_text.split()
```

```python
trump_no_stopwords = list()
for word in trump_words:
    if word not in STOP_WORDS:
        trump_no_stopwords.append(word)
```

```python
trump_freq = dict(Counter(trump_no_stopwords).most_common(50))
trump_freq
```

```python
# Generate word cloud
wordcloud = WordCloud(width=800, height=400, background_color="white").generate_from_frequencies(trump_freq)
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
